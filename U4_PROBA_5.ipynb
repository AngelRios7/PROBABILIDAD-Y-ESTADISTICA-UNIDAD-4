### **Programa 5: Problema de distribucion normal en Python**

```python
from scipy.stats import norm

# Parámetros de la distribución normal
media_peso = 70
desviacion_estandar = 4
total_estudiantes = 500

# Límites del problema
limite_inferior = 68
limite_superior = 72

# 1- Calcular la probabilidad entre 68 y 72 kg
# Se tiene que usar la función de distribución acumulada (cdf)
# Probabilidad = P(x < 72) - P(x < 68)
prob_superior = norm.cdf(limite_superior, media_peso, desviacion_estandar)
prob_inferior = norm.cdf(limite_inferior, media_peso, desviacion_estandar)

probabilidad_rango = prob_superior - prob_inferior

# 2- Calcular cuántos estudiantes representan esa probabilidad
cantidad_estudiantes = total_estudiantes * probabilidad_rango

# Resultados
print(f"La probabilidad de estudiantes que pesan entre {limite_inferior} y {limite_superior} kg es: {probabilidad_rango}")
print(f"La cantidad de estudiantes que pesan más de 68 y menos de 72 kg es: {round(cantidad_estudiantes)} estudiantes")
```
